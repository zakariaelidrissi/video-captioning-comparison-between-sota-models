#### 1.1.2 Test 2

```python
videos = os.listdir(dataset_ec_path)
print(videos)
```

```python
if not os.path.exists("results"):
    os.makedirs("results")
# Save the updated list to a JSON file
output_file_gemini = "results/ec_videos_described_gemini.json"
```

```python
videos_summaries = []
for video in videos:
    video_ec_path = dataset_ec_path + '/' + video
    summary_ec = describe_video(video_path=video_ec_path, audio=True, model="gemini")
    videos_summaries.append({
        'video_category': 'EC',
        'video_name': video,
        'summary': summary_ec
    })

with open(output_file_gemini, "w") as file:
    json.dump(videos_summaries, file, indent=4)
```

### 1.2 Medium Category

```python
dataset_mc_path = "../DatasetDivided/2-MediumCategory"
```

```python
sub_cat_mc_path = os.listdir(dataset_mc_path)
print(sub_cat_mc_path)
```

```python
if not os.path.exists("results"):
    os.makedirs("results")
# Save the updated list to a JSON file
output_file_gemini = "results/mc_videos_described_gemini.json"
```

```python
videos_summaries = []
for sub_cat in sub_cat_mc_path:
    sub_cat_path = dataset_mc_path + '/' + sub_cat
    videos = os.listdir(sub_cat_path)
    for video in videos:
        video_mc_path = sub_cat_path + '/' + video
        summary_mc = describe_video(video_path=video_mc_path, audio=True, model="gemini")
        videos_summaries.append({
            'video_category': 'MC',
            'sub_category': sub_cat,
            'video_name': video,
            'summary': summary_mc
        })

with open(output_file_gemini, "w") as file:
    json.dump(videos_summaries, file, indent=4)
```

### 1.3 Difficult Category

```python
dataset_dc_path = "../DatasetDivided/3-DifficultCategory"
```

```python
sub_cat_dc_path = os.listdir(dataset_dc_path)
print(sub_cat_dc_path)
```

```python
if not os.path.exists("results"):
    os.makedirs("results")
# Save the updated list to a JSON file
output_file_gemini = "results/dc_videos_described_gemini.json"
```

```python
videos_summaries = []
for sub_cat in sub_cat_dc_path:
    # if sub_cat == "Complex_sound_and_motion":
    print(f"******************* {sub_cat} *******************")
    sub_cat_path = dataset_dc_path + '/' + sub_cat
    videos = os.listdir(sub_cat_path)
    i = 1
    for video in videos:
        print(f"******************* {video} N {i} of {len(videos)} *******************")
        video_dc_path = sub_cat_path + '/' + video
        summary_dc = describe_video(video_path=video_dc_path, audio=True, model="gemini")
        videos_summaries.append({
            'video_category': 'DC',
            'sub_category': sub_cat,
            'video_name': video,
            'summary': summary_dc
        })
        i += 1

with open(output_file_gemini, "w") as file:
    json.dump(videos_summaries, file, indent=4)
```
